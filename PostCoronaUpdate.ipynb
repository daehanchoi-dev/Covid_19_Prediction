```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import  GridSpec
%matplotlib inline
pd.set_option('display.float_format', lambda x : '{:.2f}'.format(x))
```

```python
df_infect = pd.read_excel('../input/Global_infection.xls')
```

```python
df_infect.head()
```

```python
df = df_infect
```

```python
df_infect.drop(['filename', 'title', 'URL', 'CollectionDate', 'Channel', 'Class'], axis= 1, inplace=True)
```

```python
df.Influenza.value_counts(dropna=True)
```

```python
counts = df.Influenza.value_counts()
```

```python
useful = counts[counts>100]
useful
```

```python
print('전체 index 갯수:',len(counts))
print('뽑아 낸 유용한 index 갯수:',len(useful))
```

```python
mask = df.Influenza.isin(counts[counts<100].index)
df = df[~mask]
```

```python
df.Influenza.value_counts()
```

```python
plt.rcParams.update({'font.size': 150})
```

```python
f, ax = plt.subplots(figsize=(150, 100))
sns.countplot('Influenza', data=df)
plt.title('Influneza Values Counts', fontsize=100)
plt.xticks(rotation='80', fontsize = 75)
plt.yticks(rotation='70', fontsize= 75)
plt.xlabel('Influenza Values', fontsize=100)
plt.ylabel('Counts', fontsize=100 )
plt.tight_layout()
```
